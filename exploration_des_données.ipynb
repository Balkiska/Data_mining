```python
import pandas as pd # Por faciliter la manipulation des données csv
pd.set_option('display.max_columns', None)
import os # Pour toutes les commandes systeme 
```

```python
path = ['DataSet/games.csv',
 'DataSet/necessary_hardware.csv',
 'DataSet/open_critic.csv',
 'DataSet/social_networks.csv',
 'DataSet/tweets.csv',
 'DataSet/twitter_accounts.csv']
```

```python
df_game = pd.read_csv(path[0])
df_necessary_hardware = pd.read_csv(path[1])
df_critic = pd.read_csv(path[2])
```

```python
def keep_useful_columns(df):
    return df[['name', 'game_slug', 'price', 'release_date', 'platform', 'description', 'developer', 'publisher', 'genres', 'id_y', 'operacional_system', 'processor', 'memory', 'graphics', 'company', 'author', 'rating', 'comment', 'date']]
```

```python
def join_df(df1, df2, df3):
    concatenate = pd.merge(df1, df2, left_on='id', right_on='fk_game_id')
    concatenate = pd.merge(concatenate, df3, left_on='id_x', right_on='game_id')
    return concatenate
```

```python
# data = load_data(__PATH__)
df = keep_useful_columns(join_df(df_game, df_necessary_hardware, df_critic))
```

<h5 style="font-weight:baold;color:#AC04CD;text-align:left">Combien de jeux sont disponibles sur chaque plateforme (platform) ? Affichez les résultats sous forme de tableau.</h5>

```python


"""
Utiliser groupby(https://pandas.pydata.org/docs/reference/groupby.html#groupby) pour regrouper les données par plateforme 
Utiliser split et explode (https://pandas.pydata.org/docs/reference/api/pandas.Series.str.split.html#pandas.Series.str.split)
count(https://pandas.pydata.org/docs/reference/api/pandas.core.groupby.DataFrameGroupBy.count.html#pandas.core.groupby.DataFrameGroupBy.count) pour compter les jeux
"""


# séparer la colonne platform en listes
# platforms_split = df['platform'].str.split(',')

# explode() pour aplatir les listes dans la colonne "platform"
df_split = df.explode('platform')

# compte le nombre de jeux pour chaque plateforme
jeux_par_plateforme = df_split.groupby('platform')['name'].count().reset_index()

jeux_par_plateforme = jeux_par_plateforme.rename(columns={'name': 'nombre_jeux'})

print(jeux_par_plateforme)

```

<h5 style="font-weight:baold;color:#AC04CD;text-align:left">Quels sont les commentaires (comment) les plus longs et les plus courts dans le jeu de données ?</h5>

```python
# Calculation de la longueur du commentaire 
# ajout d'un colonne contenant la longueur du commentaire pour pouvoir comparer


df['longueur_commentaire'] = df['comment'].dropna().apply(lambda x: len(str(x)) if x is not None else 0)

# Au cas où il y a un commentaire vide
if not df.empty:

    commentaire_plus_long = df[df['longueur_commentaire'] == df['longueur_commentaire'].max()]['comment'].values[0]
    commentaire_plus_court = df[df['longueur_commentaire'] == df['longueur_commentaire'].min()]['comment'].values[0]

    print("Le commentaire le plus long est :")
    print(commentaire_plus_long)

    print("\nLe commentaire le plus courte est :")
    print(commentaire_plus_court)
else:
    print("Le DataFrame est vide. Aucun commentaire n'a été fourni.")
```
